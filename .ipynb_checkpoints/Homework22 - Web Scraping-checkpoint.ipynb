```python
# Grab the first 10 items with the original url
for i in range(9, 19):
    try:
        # Create xpath with f string
        xpath_r = f'//*[@id="main-content"]/div/ul/li[{i}]/div/div/div/div[2]/div[1]/div[1]/div[1]/div/div/h3/span/a'
        xpath_n = f'//*[@id="main-content"]/div/ul/li[{i}]/div/div/div/div[2]/div[1]/div[1]/div[3]/div/div/div/p/span[2]'
        xpath_rev = f'//*[@id="main-content"]/div/ul/li[{i}]/div/div/div/div[2]/div[1]/div[1]/div[2]/div/div/div[2]/span'
        xpath_open = f'//*[@id="main-content"]/div/ul/li[{i}]/div/div/div/div[2]/div[1]/div[2]/div/div/p/span'
        xpath_type = f'//*[@id="main-content"]/div/ul/li[{i}]/div/div/div/div[2]/div[1]/div[1]/div[3]/div/div/div/p/span[1]/a[1]/button/p'

        # Create new restaurant variable and extract the text
        restaurant = browser.find_by_xpath(xpath_r).text
        # print the restaurant
        print(f'restaurant: {restaurant}')

        # Neighborhood
        neighborhood = browser.find_by_xpath(xpath_n).text
        if neighborhood == "$$":
            'Saint Louis'
        else:
            neighborhood
        print(f'neighborhood: {neighborhood}')

        # Number of Reviews
        reviews = browser.find_by_xpath(xpath_rev).text
        print(f'reviews: {reviews}')

        # Open or closed
        openclosed = browser.find_by_xpath(xpath_open).text
        print(f'Open or closed: {openclosed}')
        
        # Type
        type_ = browser.find_by_xpath(xpath_type).text
        print(f'Type: {type_}')
        
    except Exception as e:
        print(e)

# # Add &start=10 etc. to the url to click through the next pages
# for next_ in range(10,50,10):
#     url_next = url + f'&start={next_}'
#     browser.links.find_by_href(url_next).click()
#     for i in range(19, 21):
#         restaurant = browser.find_by_xpath(xpath).text
#         print(restaurant)
```
